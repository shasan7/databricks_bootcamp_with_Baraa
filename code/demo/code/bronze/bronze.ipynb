## Read From CSV and Write Silver Table

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} â†’ bronze.{item['table']}")

    df = (
        spark.read
             .option("header", "true")
             .option("inferSchema", "true")
             .csv(item["path"])
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"bronze.{item['table']}")
    )
```
